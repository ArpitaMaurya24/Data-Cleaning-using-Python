# **Data Cleaning for Medical Appointment File**

```python
# Importing file

import pandas as pd 
import numpy as np
df=pd.read_csv(r"F:\Project\Project_2\project_2_dataset\Medical_appoint.csv")
df
```

```python
# Replacing columns value in lower case

df.columns=df.columns.str.lower()
df.info()
```

```python
# changing the datatype  and name of the columns

df['scheduledday']=pd.to_datetime(df.scheduledday)
df['appointmentday']=pd.to_datetime(df.appointmentday)
df = df.astype({'scholarship':'category','hipertension':'category','diabetes':'category','alcoholism':'category','handcap':'category','sms_received':'category','patientid':int,'gender':'category','no-show':'category'})
df=df.rename(columns={'hipertension':'hypertension','handcap':'handicap','no-show':'no_show'})
df.info()
df.describe()
```

```python
# Removing outlier in age using IQR

q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
iqr=q3-q1
upper=q3+1.5 * iqr

df = df[(df['age']>=1) & (df['age']< upper)]
df.age.describe()



```

```python
#removing rows with missing value

df=df.dropna()
```

```python
# checking each columns

df.gender.value_counts()
df.neighbourhood.value_counts()
df.scholarship.value_counts()
df.diabetes.value_counts()
df.handicap.value_counts()
df.hypertension.value_counts()
df.alcoholism.value_counts()
df.no_show.value_counts()
df.sms_received.value_counts()
```

```python
#Changing value in handicap

df['handicap']=df['handicap'].replace({2:1,3:1,4:1})
```

```python
# changing value of "no_show"

df['no_show']=df.no_show.replace({'No':0,'Yes':1})
```

```python
# Scheduled days must not be greater than appointment days

df=df[df['scheduledday']<= df['appointmentday']]
```

```python
 # formatting neighbourhood column value

df['neighbourhood']=df.neighbourhood.str.capitalize()
df['neighbourhood']=df.neighbourhood.str.strip()
```

```python
# Removing rows with multiple appointment for same patient on same day

df = df.drop_duplicates(subset=['patientid', 'appointmentday'], keep='first')
```

```python
# making the date columns proper format

df['scheduledday'] = df['scheduledday'].dt.strftime('%Y-%m-%d %H:%M:%S')
df['scheduledday'] = pd.to_datetime(df['scheduledday'])

df['appointmentday'] = df['appointmentday'].dt.strftime('%Y-%m-%d')
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
```
