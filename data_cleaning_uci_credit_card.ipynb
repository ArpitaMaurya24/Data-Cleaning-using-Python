# **Project 2: Data Cleaning For UCI Credit Card file**

```python
# Importing Data
import pandas as pd 
import numpy as np
df=pd.read_csv(r"F:\Project\Project_2\project_2_dataset\UCI_Credit_Card.csv")
```

```python
# Reviewing the datasets
df
```

```python
# Making column name in snake case format
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace('.','_')

# Replacing Columns name for more readable name
df = df.rename(columns={'pay_0':'pay_sep','pay_2':'pay_aug','pay_3':'pay_jul','pay_4':'pay_jun','pay_5':'pay_may','pay_6':'pay_apr','bill_amt1':'bill_amt_sep','bill_amt2':'bill_amt_aug','bill_amt3':'bill_amt_jul','bill_amt4':'bill_amt_jun','bill_amt5':'bill_amt_may','bill_amt6':'bill_amt_apr','pay_amt1':'pay_amt_sep','pay_amt2':'pay_amt_aug','pay_amt3':'pay_amt_jul','pay_amt4':'pay_amt_jun','pay_amt5':'pay_amt_may','pay_amt6':'pay_amt_apr'})
```

```python
# Replacing value for more clearance 

# replacing sex value from 1,2 to Male,female

df['sex'] =df.sex.replace(1,'Male')
df['sex'] =df.sex.replace(2,'Female')

# Replacing eduacation value where 1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown

mapping_edu = {1:'Graduate School',2:'University',3:'High School', 4:'Others',5:'Unknown',6:'Unknown'}
df['education']=df.education.map(mapping_edu)

# Replacing marriage value where 1 = married, 2 = unmarried, 3=others

df['marriage']=df.marriage.replace({1:'Married',2:'Unmarried',3:'Others',0:'Unknown'})

```
